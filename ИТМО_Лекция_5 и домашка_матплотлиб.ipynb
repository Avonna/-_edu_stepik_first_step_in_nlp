```python
all_cars = pd.read_csv('/content/auto-mpg.csv', 
                       delimiter=',', 
                       index_col='car name', 
                       nrows=20, 
                       decimal='.', 
                       usecols=['car name', 'horsepower', 'cylinders', 'weight'])
all_cars = all_cars.sort_values('weight')
```

```python
plt.plot('weight', 'horsepower', data = all_cars, ls = '-.', lw = '1', c = 'red', marker = '*', label = 'Ломаная со звездочками')
plt.title('Зависимость мощности автомобилей от массы') #Заголовок полотна
plt.xticks(np.arange(1600, 4700, step = 300)) #Отображаемые значения по оси $X$
plt.xlabel('Масса (кг)') #Подпись по оси X
plt.ylabel('Мощность (л/с)') #Подпись по оси Y
plt.legend(loc='lower right') #Отображение легенды в нижнем правом углу
plt.grid(c='#BEEEEE', ls='-.', lw=0.5) #Отображение сетки с заданными параметрами типа линий и их цвета
plt.show()
```

```python
x = np.linspace(-5, 5, 1000) #Массив из 100 значений на отрезке [-5, 5]
y_1 = 0.5 * x + 1 #Вычисление функции 0.5x+1 для каждого элемента x
y_2 = np.sin(x) #Вычисление функции sin() для каждого элемента x
```

```python
plt.plot(x, y_1, c='red', label='$y_1=0.5x+1$')
plt.plot(x, y_2, c='blue', label='$y_2=\sin(x)$')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.legend(loc='upper left')
plt.grid(c='#BFEFEF', ls='-', lw=0.5)
plt.show()
```

```python
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(X ** 2 + Y ** 2) - np.exp(X * Y)
```

```python
Z
```

```python
plt.contour(X, Y, Z, [0], colors=['red'])
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.xticks(np.linspace(-2, 2, 5))
plt.yticks(np.linspace(-2, 2, 5))
plt.show()
```
